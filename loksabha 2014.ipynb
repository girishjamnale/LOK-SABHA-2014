```python
total_votes=df.groupby(['Name of State/ UT']).agg({'Total Votes Polled':sum}).sum()
```

```python
total_votes
```

```python
total_votes_statewise=df.groupby(['Name of State/ UT']).agg({'Total Votes Polled': sum}).sort_values('Total Votes Polled', ascending=False).reset_index()

```

```python
plt.figure(figsize=(15,6))
plt.title('Total votes caste by State')
sns.barplot('Name of State/ UT','Total Votes Polled', data=total_votes_statewise)
plt.xticks(rotation=90)
plt.show()
```

```python
partywise_total_votes=df.groupby(['Party Name']).agg({'Total Votes Polled':sum}).sort_values( 'Total Votes Polled', ascending=False).reset_index().head(15)
```

```python
plt.figure(figsize=(15,5))
plt.title('Total votes by Party')
sns.barplot('Party Name','Total Votes Polled', data=partywise_total_votes)
plt.xticks(rotation=90)
plt.show()
```

```python
swtp=df.groupby(['Name of State/ UT','Party Name']).agg({'Total Votes Polled':sum}).sort_values(['Total Votes Polled'],ascending=False).reset_index().drop_duplicates('Name of State/ UT').reset_index()
```
